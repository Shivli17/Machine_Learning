```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn import tree
```

```python
df=pd.read_csv('bands.csv')
df
```

```python
df.columns=[
    "timestamp",  
    "cylinder number",
    "customer",
    "job number",
    "grain screened",
    "ink color",
    "proof on ctd ink",
    "blade mfg",
    "cylinder division",
    "paper type",
    "ink type",
    "direct steam",
    "solvent type",
    "type on cylinder",
    "press type",
    "press",
    "unit number",
    "cylinder size",
    "paper mill location",
    "plating tank",
    "proof cut",
    "viscosity",
    "caliper",
    "ink temperature",
    "humidity",
    "roughness",
    "blade pressure",
    "varnish pct",
    "press speed",
    "ink pct",
    "solvent pct",
    "ESA Voltage",
    "ESA Amperage",
    "wax",
    "hardener",
    "roller durometer",
    "current density",
    "anode space ratio",
    "chrome content",
    "band type"
]
df
```

```python
df.replace('?',np.nan,inplace=True)
```

```python
df.drop('timestamp',axis=1,inplace=True)
```

```python
num_cols = [
    'proof cut',
    'viscosity',
    'caliper',
    'ink temperature',
    'humidity',
    'roughness',
    'blade pressure',
    'varnish pct',
    'press speed',
    'ink pct',
    'solvent pct',
    'ESA Voltage',
    'ESA Amperage',
    'wax',
    'hardener',
    'roller durometer',
    'current density',
    'anode space ratio',
    'chrome content'
]
```

```python
df[num_cols]=df[num_cols].replace('band',np.nan)
```

```python
from sklearn.impute import SimpleImputer
df[num_cols]=df[num_cols].astype(float)
```

```python
df.info()
```

```python
mean_imp=SimpleImputer(strategy='mean')
mode_imp=SimpleImputer(strategy='most_frequent')
```

```python
df[num_cols]=mean_imp.fit_transform(df[num_cols])
cat_cols=df.select_dtypes(include='object').columns
for c in cat_cols:
    df[c]=df[c].str.lower()
df[cat_cols]=mode_imp.fit_transform(df[cat_cols])
```

```python
le=LabelEncoder()
for c in cat_cols:
    df[c]=le.fit_transform(df[c])
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
train_pred=dtc.predict(x_train)
test_pred=dtc.predict(x_test)
```

```python
print(f'Train Accuracy:{accuracy_score(train_pred,y_train)}')
print(f'Test Accuracy:{accuracy_score(test_pred,y_test)}')
```

```python
print('Train set:')
sns.heatmap(confusion_matrix(y_train,train_pred),annot=True)
plt.show()
print('Test set:')
sns.heatmap(confusion_matrix(y_test,test_pred),annot=True)
plt.show()
```

```python
with open('BANDS.txt','w') as f:
    f=tree.export_graphviz(dtc,out_file=f)
```
