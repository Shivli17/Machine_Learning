```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

## For given dataset

```python
data=pd.read_csv("C:\\Users\\HP\\Downloads\\Book 3.csv")
data
```

#### Checking if dataset is fit for linear regression

```python
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(data.X, data.Y)
```

```python
reg = linear_model.LinearRegression()
reg.fit(data[['X']], data['Y'])
```

#### Predicted value at x=7 and x=12

```python
reg.predict([[7]])
```

```python
reg.predict([[12]])
```

#### Regression equation

```python
reg.coef_
```

```python
reg.intercept_
```

```python
print("Regression equation is:",reg.coef_,"x +",reg.intercept_)
```

#### Plotting graph

```python
# for i, j in zip(data.X, data.Y):
#    plt.text(i, j+0.5, '({},{})'.format(i, j))

for x, y in zip(data.X, data.Y):
    plt.text(x, y, f'({x}, {y})', fontsize=8, ha='right', va='bottom')

plt.xlabel("X-Independent")
plt.ylabel("Y-Dependent")
plt.scatter(data.X, data.Y, color='red')
plt.plot(data.X, reg.predict(data[['X']]))
```

## For random dataset

#### generating random values

```python
import random
np.random.seed(42)  # For reproducibility
x = np.random.rand(5) 
x=x.reshape(-1, 1)
print(x)
```

```python
y = np.random.rand(5) 
y=y.reshape(-1, 1)
print(y)
```

#### Checking if dataset is fit for linear regression (values are randomly aligned).

```python
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x, y)
```

```python
reg = linear_model.LinearRegression()
reg.fit(x, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

#### Predicted value at x=7 and x=12

```python
reg.predict([[7]])
```

```python
reg.predict([[12]])
```

#### Regression equation

```python
print("Regression equation is:",reg.coef_[0][0],"x +",reg.intercept_[0])
```

#### Plotting graph

```python
for i, j in zip(x, y):
    plt.text(i, j, f'({i}, {j})', fontsize=8, ha='right', va='bottom')

plt.xlabel("X-Independent")
plt.ylabel("Y-Dependent")
plt.scatter(x, y, color='red')
plt.plot(x, reg.predict(x))
```
