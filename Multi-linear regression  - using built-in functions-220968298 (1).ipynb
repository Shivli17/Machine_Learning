```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

## For given dataset

```python
data = pd.read_csv(r"C:\Users\HP\Downloads\Multilinear dataset.csv")
data
```

```python
reg=linear_model.LinearRegression()
reg.fit(data[['X1','X2']],data.Y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
print("Regression equation is:",reg.coef_[0],"x1",reg.coef_[1],"x2 +",reg.intercept_)
```

```python
reg.predict([[3,2]])
```

```python
import seaborn as sns
sns.pairplot(data[['Y', 'X1', 'X2']])
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

sns.regplot(x='X1', y='Y', data=data, ax=ax1, ci=None, scatter_kws={'s': 100})
ax1.set_title('Partial Regression Plot for X1')

sns.regplot(x='X2', y='Y', data=data, ax=ax2, ci=None, scatter_kws={'s': 100})
ax2.set_title('Partial Regression Plot for X2')

plt.show()
```

## USING RANDOM MODULE WITHOUT INBUILT FUNCTIONS

```python
import random
np.random.seed(42)  # For reproducibility
x1 = np.random.rand(5) 
x1=x1.reshape(-1, 1)
print(x1)
```

```python
x2 = np.random.rand(5) 
x2=x2.reshape(-1, 1)
print(x2)
```

```python
y = np.random.rand(5) 
y=y.reshape(-1, 1)
print(y)
```

```python
X = np.column_stack((x1, x2))
```

```python
reg=linear_model.LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
print("Regression equation is:",reg.coef_[0][0],"x1 +",reg.coef_[0][1],"x2",reg.intercept_[0])
```

```python
reg.predict([[3,2]])
```
